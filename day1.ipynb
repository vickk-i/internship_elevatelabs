```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.info()
df.describe()
```

```python
#handling missing vals
df['director']=df['director'].fillna('Unknown')
df['cast']=df['cast'].fillna('NA')
df['country']=df['country'].fillna('Unknown')
#removing duplicates
df = df.drop_duplicates()
```

```python
nulls = df.isnull().sum()
print(nulls)
```

```python
#standardize text values
df['title'] = df['title'].str.strip().str.title()
df['cast'] = df['cast'].fillna('NA')
df['date_added']= df['date_added'].fillna('NA')
df['date_added']
```

```python
#converting date
df['date_added'] = pd.to_datetime(df['date_added'],dayfirst=True,errors='coerce')
df['date_added']
```

```python
# Rename columns to be clean and uniform
df.columns = [cols.strip().lower().replace(" ","_") for cols in df.columns]
df.columns
```

```python
# Fix data types (example: release_year as int)
df['release_year'] = df['release_year'].astype(int)
df['release_year']
```

```python
df.to_csv("netflix_titles.csv",index=False)
```
